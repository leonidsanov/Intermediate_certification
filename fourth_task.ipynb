### Задание 4

Вы решили сравнивать метрику CPA в двух группах. Размер выборки - 2350 элементов в каждой группе. Для проверки нормальности распределения на выборке в 2350 наблюдений применили , критерий Шапиро-Уилка и получили p-value, равный 0.00002, alpha = 5%.

Какой бы вывод мы могли сделать в данном случае?

В этом случае какой статистический критерий для проверки первоначальной гипотезы тут лучше всего подойдёт и почему?


Нулевая гипотеза теста Шапиро-Уилка заключается в том, что популяция распределена нормально. Таким образом, если значение p-value меньше выбранного $\alpha$, то нулевая гипотеза отклоняется и есть доказательства того, что проверяемые данные распределены не нормально. С другой стороны, если значение p-value больше выбранного $\alpha$, то нулевая гипотеза (о том, что данные получены из нормально распределенной совокупности) не может быть отклонена.

В данном случае p-value < $\alpha$, поэтому принимаем альтернативную гипотезу о том, что выборка *не принадлежит нормальному распределению*.

Cost per action (CPA) - цена за целевое действие - метрика, относящаяся к количественному непрерывному типу данных. Кроме того в данном случае данные не принадлежат нормальному распределению, а выборки являются независимыми, поэтому для проверки первоначальной гипотезы лучше всего подойдет критерий Манна-Уитни.
