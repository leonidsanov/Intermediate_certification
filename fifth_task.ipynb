### Задание 5
Мы провели АБ-тест на увеличение average timespent per user. По итогам тестирования мы получили следующие данные. Является ли результат статистически значимым с уровнем доверия 80%? Какую версию мы выкатим на продакшн?

A) Средняя - 360, отклонение - 40, количество - 9802

B) Средняя - 352, отклонение - 58, количество - 9789

```python
# импорт необходимых библиотек
import plotly.graph_objects as go # Для построения графиков
from scipy import stats # Для расчёта ститистик
import numpy as np # Для работы с массивами данных
```

Average time spent per user (сколько в среднем времени пользователь проводит в сервисе за период) является количественным непрерывным типом данных.

```python
# Вводим исходные данные по выборкам
sample_1_mean, sample_2_mean = 360, 352 # среднее значение по группам
sample_1_se, sample_2_se = 40, 58 # стандартное отклонение по группам
sample_1_size, sample_2_size = 9802, 9789 # количество наблюдений по группам
# Генерируем нормальное распределение на основе введенных данных
norm_1 = stats.norm(loc=sample_1_mean, scale=sample_1_se).rvs(size=sample_1_size)
norm_2 = stats.norm(loc=sample_2_mean, scale=sample_2_se).rvs(size=sample_2_size)
```

В данном случае нет понимания того, является ли распределение нормальным, поэтому мы будем генерировать данные на основе известных законов распределения.

```python
fig = go.Figure()
fig.add_trace(go.Histogram(x=norm_1, name='norm_1'))
fig.add_trace(go.Histogram(x=norm_2, name='norm_2'))

# Наложение обеих гистограмм
fig.update_layout(barmode='overlay')
# Уменьшите непрозрачность, чтобы увидеть обе гистограммы.
fig.update_traces(opacity=0.5)
fig.show()
```

```python
# Считаем статистику по выборкам
tvalue, pvalue = stats.ttest_ind(norm_1, norm_2, equal_var=False, alternative='two-sided')
md = sample_1_mean - sample_2_mean
se = md / tvalue

print(f"Difference between samples: {md}, standart error: {se}, p_value: {pvalue}")

if pvalue < .20:
    print(f"Sample # {1 if md > 0 else 2} mean is greater")
else:
    print("No significant difference")
```

```python
#Что если распределение не нормальное?
# Генерируем экспоненциальное распределение на основе введенных данных
expon_1 = stats.expon(loc=sample_1_mean, scale=sample_1_se).rvs(sample_1_size)
expon_2 = stats.expon(loc=sample_2_mean, scale=sample_2_se).rvs(sample_2_size)
```

```python
fig = go.Figure()
fig.add_trace(go.Histogram(x=expon_1, name='expon_1'))
fig.add_trace(go.Histogram(x=expon_2, name='expon_2'))

# Наложение обеих гистограмм
fig.update_layout(barmode='overlay')
# Уменьшите непрозрачность, чтобы увидеть обе гистограммы.
fig.update_traces(opacity=0.75)
fig.show()
```

```python
# Считаем статистику по выборкам
mwvalue, pvalue = stats.mannwhitneyu(expon_1, expon_2)

md = sample_1_mean - sample_2_mean
se = md / mwvalue

print(f"Difference between samples: {md}, standart error: {se}, p_value: {pvalue}")

if pvalue < .20:
    print(f"Sample # {1 if md > 0 else 2} mean is greater")
else:
    print("No significant difference")
```

По результатам обоих тестов среднее значение Average time spent per user у первой выборки больше, из чего можно сделать вывод о том, что "выкатить на продакшн" необходимо версию А.
